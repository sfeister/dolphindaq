# PVAccess Dropped Frames Test

How many frames are we dropping from our electron trace if we use the callback methods? (a.k.a. asynchronous pvget)

```python
import numpy as np
import matplotlib.pyplot as plt

import time
import datetime
from time import sleep

import pvaccess as pva
```

## System Configuration

### Basic settings like repetition rate

```python

proton_dt = pva.Channel("PROTON:dt:set", pva.CA)
electron_dt = pva.Channel("ELECTRON:dt:set", pva.CA)
laser_dt = pva.Channel("LASER:powers:dt:set", pva.CA)
pulse_rep = pva.Channel("PULSEGEN:reprate:set", pva.CA)

proton_dt.put(5.0e-6) # seconds
electron_dt.put(5.0e-6) # seconds
laser_dt.put(5) # microseconds
pulse_rep.put(50) # Hz
```

## Capturing Frames

```python
etrace_PVA = pva.Channel("ELECTRON-DAQ:trace", pva.PVA)
ptrace_PVA = pva.Channel("PROTON-DAQ:trace", pva.PVA)
```

```python
oldId = 0
def echo(pvobj):
    global oldId
    newId = pvobj.get()["uniqueId"]
    print(newId - oldId)
    oldId = newId

def error(code):
    raise Exception("Returned error code: {}".format(code))

etrace_PVA.asyncGet(echo, error, 'field(uniqueId,timeStamp)')
```

```python
etrace_PVA.stopMonitor()
```

```python
etrace_PVA.monitor(echo, 'field(value,uniqueId,timeStamp)')
```

```python
etrace_PVA.subscribe('echo', echo)
```
